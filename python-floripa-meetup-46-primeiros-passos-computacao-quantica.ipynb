# A Computação Quântica

-   explora os efeitos da mecanica-quantica
-   algoritimos do início da década de 90
    - Shor's (quebra criptografia) e Groover (busca)
-   lida basicamente com manipulação de estados quanticos
-   o estado de um sistema quantico é representado por um vetor em um espaço vetorial (linear) complexo (Hilbert space)
-   você precisa trabalhar com números complexos
    -   parte imaginária e uma parte real (parte imaginaria numero
        imaginário raiz quadrada de -1)
-   o hardware influencia o comportamento
-   potencial de reduzir drasticamente tempo de execução e consumo de energia
-   a supremacia quântica
-   quantum bit (**qubit**) vs. bit clássico (**cbit**)

<a id="org0131c54"></a>

# Computação reversível e não reversível

-   na computação classica as operações "identidade" e "negação" são
    reversíveis
-   na computação classica as operações "constante-0" e "constante-1"
    não são reversíveis
-   na computação quantica só existem operações reversíveis, e todos os
    operadores são inversão deles próprios

<a id="org0ec1742"></a>

# Conceitos da mecânica quântica

-   **superposição** é quando há uma combinação de dois estados quanticos
    normalizados, formando um estado quantico válido
    -   um qubit pode ser 0 e 1 ao mesmo tempo

-   **entrelaçamento** é quando dois ou mais qubits tem seus estados um
    dependente do outro

-   **medida** é quando se transforma o estado quantico, armazenado em um
    sistema quantico, em uma informação clássica 
    -   seja o resultao 0 ou 1 esta saida é probabilística
    -   durante a medida o valor do qubit colapsa em um valor 0 ou 1
    -   geralmente utilizado no fim de um conjunto de operações em um
        <span class="underline">circuito</span> quantico

<a id="orgc4f2518"></a>

# O qubit (quantum bit)

-   menor unidade de medida de um computador quantico
-   bit classico armazena 0 e 1
-   bit classico quando medido o valor não é alterado, permanece o mesmo
-   qubit armazena qualquer estado entre 0 e 1
-   qubit é um sistema de duas dimensões
-   diferente de um bit classico, o estado de um qubit não pode ser
    medido sem alterá-lo

<a id="orgcd3d529"></a>

# Representação do Qubit

- qubit representado por um vetor:
$\begin{pmatrix}
a\\
b\\
\end{pmatrix}$
onde _a_ e _b_ são números complexos e: $\mid\mid a \mid\mid^2 + \mid\mid b \mid\mid^2 = 1$

- se um qubit tem um valor 
$\begin{pmatrix}
a\\
b\\
\end{pmatrix}$
então a probabilidade dele ser **0** é $\mid\mid a \mid\mid^2$ e a probabilidade de ele ser **1** é de $\mid\mid b \mid\mid^2$ 

- $\begin{pmatrix}
1\\
0\\
\end{pmatrix}$
tem 100% de chances de colapsar para **0**, já
$\begin{pmatrix}
0\\
1\\
\end{pmatrix}$
tem 100% de chances de colapsar para **1**

- $\begin{pmatrix}
\sqrt{\frac{1}{2}}\\
\sqrt{\frac{1}{2}}\\
\end{pmatrix}$
tem 50% de chances de tanto colapsar para **0** ou para **1** pois $\mid\mid\sqrt{\frac{1}{2}}\mid\mid^2=\frac{1}{2}$

- notação *bra-ket* ou *Dirac*
$\begin{pmatrix}
1\\
0\\
\end{pmatrix}=|0>
\begin{pmatrix}
0\\
1\\
\end{pmatrix}=|1>$

# Circuito simples

```python
qc = QuantumCircuit(1,1)

# [0, 1, 2]
qc.x(0)
qc.measure(0,0)
qc.draw(output='mpl')
```

# Porta NOT

```python
qc = QuantumCircuit(2, 2)

# Entradas possiveis
#qc.h(0)

# Processamento
qc.cx(0,1)

# Saida
qc.measure([0,1], [0,1])
```

```python
#qc.draw()
qc.draw(output='mpl')
```

# Porta AND

```python
qc = QuantumCircuit(3, 3)

# Entrada manual

#qc.x(0)
#qc.x(1)

# Todas as entradas possíveis
qc.h(0)
qc.h(1)

# Processamento
qc.barrier()

qc.ccx(0,1,2)

qc.barrier()

# Saida
qc.measure([0, 1,2], [0,1,2])
```

```python
qc.draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')

job = execute(qc, backend)

result = job.result()

output_counts = result.get_counts()

plot_histogram(output_counts)
```

# Porta OR

```python
qc = QuantumCircuit(3, 3)

qc.h(0)
qc.h(1)

qc.barrier()

qc.x(0)
qc.x(1)

qc.ccx(0, 1, 2)

qc.x(0)
qc.x(1)
qc.x(2)

qc.barrier()

qc.measure([0,1,2], [0,1,2])
qc.draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')

job = execute(qc, backend, shots=)

result = job.result()

output_counts = result.get_counts()

plot_histogram(output_counts)
```
